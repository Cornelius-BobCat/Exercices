```python
def afficher_plateau(plateau):
  for ligne in plateau:
    print("|".join(ligne))
    print("-"*5) 

def verifier_victoire(plateau, joueur_actuel):

  # check lignes
  for ligne in plateau:
    if ligne.count(joueur_actuel) == 3:
      return True
    
  # check les colonnes
  for col in range(3):
    if plateau[0][col] == plateau[1][col] == plateau[2][col] == joueur_actuel:
      return True

  # check diagonale
  if plateau[0][0] == plateau[1][1] == plateau[2][2] == joueur_actuel:
    return True
  if plateau[0][2] == plateau[1][1] == plateau[2][0] == joueur_actuel:
    return True
  
  return False
    
  


def game():
  plateau = [[" " for _ in range(3)] for _ in range(3)] # [["X"," "," "],[" "," "," "],[" "," "," "]]
  joueur_actuel = "X"

  while True:
    afficher_plateau(plateau)
    print(f"C'est tour du joueur {joueur_actuel} ")
    while True:
      ligne = int(input("Choissisez la ligne (0, 1, 2) : "))
      if ligne in range(3):
        break
      else:
        print("rentre un numéro valide idiot!")
    while True:
      colonne = int(input("Choissisez la colonne (0, 1, 2) : "))
      if colonne in range(3):
        break
      else:
        print("rentre un numéro valide idiot!")

    #vérifier si la case est vide 
    if plateau[ligne][colonne] == " ":
      plateau[ligne][colonne] = joueur_actuel
      # check si vainqueur
      if verifier_victoire(plateau, joueur_actuel):
        print(f"Féliciation {joueur_actuel} vous avez gagné!")
        break

      #match nul 1
      
      if all([all([cellule != ' ' for cellule in ligne]) for ligne in plateau ]):
        afficher_plateau(plateau)
        print("Match NUL !! ")
        break

      joueur_actuel = "O" if joueur_actuel == "X" else "X"
    else:
      print("cette case est déjà occupée !")


game()
```
